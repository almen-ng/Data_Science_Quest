# Introduction to Machine Learning Fundamentals

**Artificial Intelligence:** The effort to automate intellectual tasks normall performed by humans; A.I. tries to simulate some intellectual human behavior

1950s -> Artificial Intelligence was defined as a predefined set of rules

Examples include:
* Tic Tac Toe
* Chess
* Pac Man 
    * path-finding (considered AI)

**Machine Learning:** figures out the rules for us and needs data in order to do so. 
* Give input data + what should it be (expected output data) = generates rules for us 
    * Rather than giving the program rules, an algorithm finds the rules for us 
* Might not get 100% accuracy
* GOAL: Raise accuracy as high as possible

**Neural Networks:** A form of machine learning that uses a layered representation of data 
* Multistage Information Extraction Process

Data is the most important part of machine learning and neural networks 

Input: Features - the information; give to the model\
Output: Label - what we look to predict

Give features and labels to a model to start training. It comes up with rules such that if we just give features to the model in the future it would give a good estimate of what the output should be.
